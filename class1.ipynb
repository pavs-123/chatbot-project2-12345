```python
# Example 1: Basic TypedDictfrom typing import TypedDictclass User(TypedDict):    name: str    age: int    is_active: booluser: User = {"name": "Alice", "age": 30, "is_active": True}print(f"User: {user}")
```

```python
# Example 2: TypedDict with Optional fieldsfrom typing import TypedDict, Optionalclass Employee(TypedDict):    name: str    employee_id: int    department: Optional[str] = None  # Optional field with default    salary: Optional[float] = Noneemp1: Employee = {"name": "Bob", "employee_id": 101}emp2: Employee = {"name": "Carol", "employee_id": 102, "department": "Engineering", "salary": 75000.0}print(f"Employee 1: {emp1}")print(f"Employee 2: {emp2}")
```

```python
# Example 3: Nested TypedDictfrom typing import TypedDict, Listclass Address(TypedDict):    street: str    city: str    zip_code: strclass Person(TypedDict):    name: str    age: int    address: Addressclass Company(TypedDict):    name: str    employees: List[Person]address: Address = {"street": "123 Main St", "city": "Boston", "zip_code": "02101"}person: Person = {"name": "David", "age": 35, "address": address}company: Company = {    "name": "TechCorp",    "employees": [person]}print(f"Company: {company}")print(f"Employee lives in: {person['address']['city']}")
```

```python
# Example 4: TypedDict with total=False (all fields optional)from typing import TypedDictclass Config(TypedDict, total=False):    debug: bool    max_connections: int    timeout: floatconfig1: Config = {}  # Empty dict is validconfig2: Config = {"debug": True}  # Partial fieldsconfig3: Config = {"debug": False, "max_connections": 100, "timeout": 30.5}print(f"Config 1: {config1}")print(f"Config 2: {config2}")print(f"Config 3: {config3}")
```

```python
# Example 5: TypedDict with Union typesfrom typing import TypedDict, Union, Listclass Product(TypedDict):    product_id: int    name: str    price: Union[int, float]    tags: List[str]product: Product = {    "product_id": 1,    "name": "Laptop",    "price": 999.99,    "tags": ["electronics", "computers"]}print(f"Product: {product}")
```
